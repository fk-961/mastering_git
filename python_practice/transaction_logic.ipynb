# Transaction logic

```python
transactions = [
    {"amount" : 50, "category" : "food"},
    {"amount" : 120, "category" : "transport"},
    {"amount" : 80, "category" : "food"},
]
```

## Useful functions:
- total spending
- get transactions of a specific category
- get the biggest transaction
- get the spending by each category

```python
def total_spending(transactions : list) -> float:
    total = 0
    for transaction in transactions:
        total += transaction["amount"]
    return total

def get_categories(transactions : list) -> list:
    categories = set()
    for transaction in transactions:
        categories.add(transaction["category"])
    return sorted(list(categories))

def filter_category(
    transactions : list,
    category : str
) -> list:
    if category not in get_categories(transactions):
        raise ValueError("Category not found !")
    else:
        matching_transactions = []
        for transaction in transactions:
            if transaction["category"] == category:
                matching_transactions.append(transaction)
        return matching_transactions
    
def biggest_transaction(transactions : list) -> float:
    amounts = []
    for transaction in transactions:
        amounts.append(transaction["amount"])
    return max(amounts)

def spending_by_category(transactions):
    spending = {}
    for transaction in transactions:
        if transaction["category"] in spending:
            spending[transaction["category"]] += transaction["amount"]
        else:
            spending[transaction["category"]] = transaction["amount"]
    return spending
```

```python
def run_tests(transactions : list) -> None:
    print(f"Total spending: {total_spending(transactions)}")
    print(f"Get food transactions: {filter_category(transactions, "food")}")
    print(f"biggest_transaction: {biggest_transaction(transactions)}")
    print(f"Spending by category: {spending_by_category(transactions)}")
```

```python
run_tests(transactions)
```

# Work with CSV

```python
import csv
```

```python
def load_transactions_csv(filepath : str) -> list:
    with open(filepath, newline = '') as f:
        reader = csv.DictReader(f)
        transactions_csv = []
        for row in reader:
            row['amount'] = int(row['amount'])
            transactions_csv.append(row)
    return transactions_csv
```

```python
transactions_csv = load_transactions_csv("./transactions.csv")
```

```python
run_tests(transactions_csv)
```
