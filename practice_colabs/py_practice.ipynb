## List Comprehension
The example below is the same as above just in one line

```python
# nested list comprehension

perms = [
    (i,j,k) for i in range(2) for j in range(2) for k in range(2)
]

perms
```

```python
[nice for nice in perms if sum(nice) != 2]
```

```python
# flattening lists of list
mat = [[1,2,3],[4,5,6],[7,8,9]]
[x for i in mat for x in i]
```

```python
arr = [2,3,6,6,4]

for i in arr:
    if i == max(arr)-1:
        print(i)
```

```python
runner = 0

for x in arr:
    if x < max(arr) and x > runner:
        runner = x
        
print(x)
```
