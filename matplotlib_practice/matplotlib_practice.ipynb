```python
import matplotlib.pyplot as plt
import numpy as np
```

Every graph is composed of a figure and one or more axes.

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5],[1,4,2,3,2])
plt.show()
```

## Components of a Figure
- plt.figure
---
- ax.set_title
- ax.legend
- ax.xaxis
- ax.set_xlabel
---
- ax.xaxis.set_major_locator
- ax.xaxis.set_minor_locator
- ax.xaxis.set_major_formatter
- ax.xaxis.set_minor_formatter
(same for yaxis)
---
- ax.grid
- ax.spines
---
- ax.scatter
- ax.plot


```python
fig, ax = plt.subplots()

ax.set_title("Reste de la division par n pour les entiers")

for n in range(1,6):
    ax.scatter(
        [i for i in range(10)],
        [i%n for i in range(10)],
        label = "n="+str(n)
    )
    
ax.xaxis.set_ticks([i for i in range(10)])
ax.xaxis.set_label_text("Entier")

ax.yaxis.set_ticks([i for i in range(10)])
ax.yaxis.set_label_text("Reste")

ax.grid()
ax.legend()

plt.show()
```

A figure can contain multiple axes with each their propreties by adding arguments to `plt.subplots(2, 2)` or using `plt.subplot_mosaic` for example.
